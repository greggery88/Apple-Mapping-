```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import geoplot as gplt
import geopandas as geopd
import geoplot.crs as gcrs
```

```python
cd = geopd.read_file("https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/US-counties.geojson")
cd = cd[cd["STATE"]=="23"].reset_index()
cd = cd.drop(columns=["STATE", "index", "id", "GEO_ID", "CENSUSAREA", "COUNTY", "LSAD"])
cd = cd.rename(columns={"NAME":"county"})
cd
```

```python
t = gplt.polyplot(cd, projection=gcrs.AlbersEqualArea())
```
